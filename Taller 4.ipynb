Dando continuidad con la primera entrega del proyecto, en esta oportunidad el estudiante debe realizar las siguientes validaciones utilizando la sentencia condicional IF.

Si el empleado es mayor de 55 años disfrutará de un bono de prepensión correspondiente al 5% de su sueldo básico.

Si el empleado es casado y tiene hijos se le otorgará un paseo cada diciembre

Si el sueldo básico está entre 1000000 y 1500000 tendrá una comisión del 2% sobre el valor del sueldo; Si el sueldo básico está entre 1500001 y 2000000 tendrá una comisión del 5% sobre el valor del sueldo; para todos los demás casos no habrá comisión.

Si el empleado trabajó más de 20 días al mes y su sueldo es menor a 1000000 tendrá derecho a un bono de alimentación.

```python
nombres = input ("Escriba su nombre: ")
apellidos = input ("Escriba su apellido: ")
edad = int(input ("¿Cual es su edad?: "))
estadoCivil = int(input ("ESTADO CIVIL: (1).Soltero (2).Union Libre (3).Casado (4).Viudo "))
numeroHijos = int(input ("¿Cuantos hijos tienes? "))
sueldoBasico = float(input ("Digite su sueldo: "))
diasLaborados = input ("Dias Laborados: ")
print (" ")

def bono(ed):
  if ed >= 55:
    bono=sueldoBasico * 0.5
    print("El bono de prepension para " + nombres + " " + apellidos +  " es {:,.0f}".format(bono))
  else:
    print("NO TIENE DERECHO A BONO")

def paseo (civil,hijos):
  if civil == 3 and hijos > 0:
    print("El empleado " + nombres + " " + apellidos + " tiene derecho a paseo cada diciembre")
  else:
    print("El empleado " + nombres + " " + apellidos + " NO tiene derecho a paseo")

def comision(sueldo):
 if sueldo >= 1000000 and sueldo <= 1500000:
  comision = sueldo * 0.2
  print("La comision para " + nombres + " " + apellidos +  " es {:,.0f}".format(comision))
 elif sueldo >= 1500001 and sueldo <= 2000000:
  comision = sueldo * 0.5
  #Para separar variables utilizar , ya que con + genera error por el tipo de variable
  print("La comision para " , nombres , " " , apellidos ,  " es {:,.0f}".format(comision))
 else:
  print("NO TIENE COMISION")

def alimentacion(dias,sueldo):
  if dias > "20" and sueldo < 1000000:
    print("El empleado " , nombres , " " , apellidos ,  " tiene derecho a bono de alimentacion ")
  else:
    print("No tiene derecho a bono de alimentacion")

bono(edad)
paseo(estadoCivil,numeroHijos)
comision(sueldoBasico)
alimentacion(diasLaborados,sueldoBasico)
```
